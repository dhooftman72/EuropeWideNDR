```python
import arcpy
with arcpy.EnvManager(overwriteOutput=True):
   # MODIS NDVI files with calculation of C factors loop following Kleijn et al. 2000; INVERSED to % Cover
   Numbers = ["001","017","033","049","065","081","097","113","129","145","161","177","193","209","225","241","257","273","289","305","321","337","353"]
   Years =   ["2018"] 
   Folder = r"**********"   
   FolderOut =  "********" 
   InBase = Folder + "MOD13Q1.061__250m_16_days_NDVI_doy" 
   for day in range(1,len(Numbers)):
       for year in range(0,1):
           Message = Years[year] + "_" + Numbers[day] + "   "
           print(Message, end="\r")          
           f = ["WorkRast","ValueRast", "Value32","Lookup_Rast", "Minus_Rast", "Divide_Rast", "Times_Rast", "Exp_Rast",\
               "ConvertRast","ProjectRaster","SampleRast","ClipRast","NullRast"]
           for i in f:
               arcpy.Delete_management(i)
                
           # Name file
           InFile = InBase + Years[year] + Numbers[day]+ "_aid0001.tif"
           OutFileSoilCover =  FolderOut+ "CRaster_" + Numbers[day] + ".tif"
           arcpy.management.CopyRaster(InFile, "WorkRast", '',None, "-3000", "NONE", "NONE", '',\
                                                "NONE", "NONE", "GRID", "NONE", "CURRENT_SLICE", "NO_TRANSPOSE")
           # Make sure values are between 0-1   
           out_raster = arcpy.sa.Con("WorkRast", 0, "WorkRast", "VALUE < 0")
           out_raster.save("ValueRast")
           arcpy.management.AddField("ValueRast", "Value32", "DOUBLE", 10, 5, None, '', "NULLABLE", "NON_REQUIRED", '') 
           arcpy.management.CalculateField("ValueRast", "Value32", "!Value!/10000", "PYTHON3", '', "TEXT", "NO_ENFORCE_DOMAINS")
           #Transition from NDVI to cover managment factors follow Kleijn et al. 2000
           out_raster = arcpy.sa.Lookup("ValueRast", "Value32"); out_raster.save(r"Lookup_Rast")
           out_raster = arcpy.sa.Minus(1, "Lookup_Rast"); out_raster.save(r"Minus_Rast")
           out_raster = arcpy.sa.Divide("Lookup_Rast", "Minus_Rast"); out_raster.save(r"Divide_Rast")
           out_raster = arcpy.sa.Times(-2, "Divide_Rast"); out_raster.save(r"Times_Rast")
           out_raster = arcpy.sa.Exp("Times_Rast"); out_raster.save(r"Exp_Rast")        
           # Make it cover not managment factor!!
           out_raster = arcpy.sa.Minus(1, "Exp_Rast"); 
           out_raster.save("ConvertRast") 
           arcpy.management.ProjectRaster(in_raster="ConvertRast",out_raster="ProjectRaster",
              out_coor_system='PROJCS["GRS_1980_IUGG_1980_Lambert_Azimuthal_Equal_Area",GEOGCS["GCS_GRS_1980_IUGG_1980",DATUM["D_unknown",SPHEROID["GRS80",6378137.0,298.257222101]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]],PROJECTION["Lambert_Azimuthal_Equal_Area"],PARAMETER["false_easting",4321000.0],PARAMETER["false_northing",3210000.0],PARAMETER["central_meridian",10.0],PARAMETER["latitude_of_origin",52.0],UNIT["Meter",1.0]]',
              resampling_type="NEAREST",cell_size="373.230919188889 373.230919188889",geographic_transform=None,Registration_Point=None,
              in_coor_system='GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137.0,298.257223563]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]]',
              vertical="NO_VERTICAL")   
           # arcpy.management.Resample(in_raster="ProjectRaster", out_raster="SampleRast",cell_size="100 100",resampling_type="NEAREST")
           arcpy.management.Clip(in_raster="ProjectRaster",
              rectangle="-18.1696376794353 27.6388187403927 40.2046165795915 71.1881103518114",
              out_raster="ClipRast",in_template_dataset="GeneralShapeBuffered500m",nodata_value="-999",
                 clipping_geometry="ClippingGeometry", maintain_clipping_extent="MAINTAIN_EXTENT")                  
           #out_raster = arcpy.sa.SetNull(in_conditional_raster="IsNullEurope100mRasterBuffered.tif",in_false_raster_or_constant="ClipRast",
                          #where_clause="Value = 1")
           #out_raster.save("NullRast")
           out_raster = arcpy.sa.Con(in_conditional_raster="ClipRast" ,in_true_raster_or_constant=1,
                        in_false_raster_or_constant="ClipRast", where_clause="VALUE > 1")
           out_raster.save(OutFileSoilCover)

# Afterwards manually average the rasters per year (using cell statistics)
```

```python
# USED TRY 2

# Delineating watersheds TRY 2; First clip the accumulation and direction raster in watershed parts and
# then run a point for the respective watershed only. Partly Level 3 watersheds are used were level 4 is too fine
# i.e., waterways (e.g. the donau) cross multiple sheds and there are measurment points on such (I checked)
# For Level 3 watersheds 852, 976, 1076 & 1302, the watersheds are split into level 3, since those are true watersheds

import arcpy
import time

ShedDirectory = "*********"       
PointsNames = "EioNetRiverConcentrationsP15_N10Cleaned"
expression= "DH_ID > 1745"
count = 0
time_needed = []
f = ["SelectedPoint","SnapSelected","RastPoly","Watersh_Rast","WaterShedstmp","SelectedPoly","Selectedtmp"]
for x in f:
    arcpy.Delete_management(x)
with arcpy.EnvManager(outputCoordinateSystem='PROJCS["GRS_1980_IUGG_1980_Lambert_Azimuthal_Equal_Area",GEOGCS["GCS_GRS_1980_IUGG_1980",DATUM["D_unknown",SPHEROID["GRS80",6378137.0,298.257222101]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]],PROJECTION["Lambert_Azimuthal_Equal_Area"],PARAMETER["false_easting",4321000.0],PARAMETER["false_northing",3210000.0],PARAMETER["central_meridian",10.0],PARAMETER["latitude_of_origin",52.0],UNIT["Meter",1.0]]', 
        cellSize= 25, overwriteOutput=True):
    with arcpy.da.SearchCursor(PointsNames, ["DH_ID"], where_clause=expression) as cursor: 
        for row in cursor:
            start_time = time.time()
            count = count + 1
            exp = '"DH_ID" = ' + str(row[0])
            Message = "Running DH_ID_" + str(row[0]) + ", which is the "+ str(count) + "th watershed run; the previous one needed " + str(time_needed) + " seconds"
            print(Message, end="\r")  
            arcpy.analysis.Select(in_features="EioNetRiverConcentrationsP15_N10Cleaned",
                 out_feature_class="SelectedPoint",where_clause=exp)
            # It is allowed the near distance calcuated above off, plus 25 meters (one pixel) to cover a small alignment errors. 
            #Note it need to be pre-snapped with the near tool
            
            File = arcpy.SearchCursor("SelectedPoint")
            for rw in File:
                ShedNum = rw.getValue("ShedToRun") 
            AccRaster = ShedDirectory + "AccumulationRaster_" + str(ShedNum) + ".tif"    
            DirRaster = ShedDirectory + "DirectionRaster_" + str(ShedNum) + ".tif"   
            
            File = arcpy.SearchCursor("SelectedPoint")
            for rw in File:
                valueOfField = rw.getValue("NEAR_DIST") 
            valueOfField = valueOfField + 26
            exp_dis = str(valueOfField)   
            out_raster = arcpy.sa.SnapPourPoint(in_pour_point_data="SelectedPoint",
                 in_accumulation_raster=AccRaster,snap_distance=exp_dis,pour_point_field="DH_ID")
            out_raster.save("SnapSelected")
            with arcpy.EnvManager(parallelProcessingFactor="16"):
                out_raster = arcpy.sa.Watershed(in_flow_direction_raster=DirRaster,
                         in_pour_point_data="SnapSelected",pour_point_field="Value")
                out_raster.save("Watersh_Rast")
            arcpy.conversion.RasterToPolygon(in_raster="Watersh_Rast",out_polygon_features="RastPoly",
                    simplify="NO_SIMPLIFY",raster_field="Value",create_multipart_features="MULTIPLE_OUTER_PART",max_vertices_per_feature=None)
            arcpy.management.AddField(in_table="RastPoly",field_name="DH_ID",field_type="LONG",
                    field_precision=None,field_scale=None,field_length=None,field_alias="",field_is_nullable="NULLABLE",
                    field_is_required="NON_REQUIRED",field_domain="")
            arcpy.management.CalculateField(in_table="RastPoly",field="DH_ID", expression=row[0],
                    expression_type="PYTHON3",code_block="",field_type="TEXT",enforce_domains="NO_ENFORCE_DOMAINS")
            arcpy.management.JoinField(in_data="RastPoly",in_field="gridcode",join_table="SelectedPoint",join_field="DH_ID",
                    fields="OBJECTID;waterBodyI;lon;lat;NO2_mg_l_;P_mg_l_;NO3_mg_l_;NH4_mg_l_;Total_N_mg;PH;NO2_Num;P_Num;NO3_num;NH4_num;PH_num;TotalN_avg;NEAR_DIST;ShedToRun",
                    fm_option="NOT_USE_FM",field_mapping=None)
            arcpy.conversion.RasterToPoint(in_raster="SnapSelected",out_point_features="SelectedPoly",raster_field="Value")
            arcpy.management.AddField(in_table="SelectedPoly",field_name="DH_ID",field_type="LONG",
                    field_precision=None,field_scale=None,field_length=None,field_alias="",field_is_nullable="NULLABLE",
                    field_is_required="NON_REQUIRED",field_domain="")
            arcpy.management.CalculateField(in_table="SelectedPoly",field="DH_ID", expression=row[0],
                    expression_type="PYTHON3",code_block="",field_type="TEXT",enforce_domains="NO_ENFORCE_DOMAINS")
            # Merge all polygons into one layer, note it will have overlapping watersheds          
            if count == 1:
             # generate the merging file                                        
                arcpy.management.CopyFeatures("RastPoly", "*****WaterSheds_Merge.shp" , '', None, None, None)  
                arcpy.management.CopyFeatures("SelectedPoly", "*******SelectedPoints_Merge.shp" , '', None, None, None)  
            else:  
             # Join with other watersheds               
                arcpy.management.Merge(inputs=r"**********WaterSheds_Merge.shp;RastPoly",
                        output="WaterShedstmp",field_mappings= *********,
                        add_source="NO_SOURCE_INFO")                                                                    
                arcpy.Delete_management("********WaterSheds_Merge.shp") 
                arcpy.management.CopyFeatures(in_features="WaterShedstmp",out_feature_class="****WaterSheds_Merge.shp",
                        config_keyword="",spatial_grid_1=None,spatial_grid_2=None,spatial_grid_3=None)
                
                arcpy.management.Merge(inputs="*******SelectedPoints_Merge.shp;SelectedPoly",
                        output="Selectedtmp",field_mappings=*******,
                        add_source="NO_SOURCE_INFO")
                arcpy.Delete_management("*********SelectedPoints_Merge.shp") 
                arcpy.management.CopyFeatures(in_features="Selectedtmp",out_feature_class="*********SelectedPoints_Merge.shp",
                        config_keyword="",spatial_grid_1=None,spatial_grid_2=None,spatial_grid_3=None)
                         
            f = ["SelectedPoint","SnapSelected","RastPoly","Watersh_Rast","WaterShedstmp","SelectedPoly","Selectedtmp"]
            for x in f:
                arcpy.Delete_management(x)
            time_needed = (time.time() - start_time)  
        print("Ready")        
```
